# Множественная линейная регрессия (multiple linear regression)
Датасет взят отсюда: https://www.kaggle.com/datasets/hellbuoy/car-price-prediction

```python
import pandas as pnd
from sklearn import linear_model
from sklearn.metrics import r2_score
import numpy as np
from matplotlib import pyplot as plt
data = pnd.read_csv('CarPrice.csv') #загружаем датасет из csv-файла
print(list(data.columns)) #выводим список заголовков датасета
```

<br>Для начала попробуем использовать один из параметров, который должен влиять на стоимость, чтобы построить одиночную линейную регрессию.
<br>Предполагаем, что цена напрямую зависит от мощности двигателя, т.е. price = a*horsepower+b, a, b - неизвестны

```python
a = data['price'].corr(data['horsepower'])*data['price'].std()/data['horsepower'].std()
#y = ax+b, a = corr(x,y)*std(y)/std(x)
print('a =', a)
```

```python
b = data['price'].mean() - a*data['horsepower'].mean()
#b = mean(y) - mean(x)*a
print('b = ', b)
print('y = {:.2f}x{:+.2f}'.format(a,b))
```

```python
plt.plot(data['horsepower'], data['price'], 'yo',data['horsepower'], a*data['horsepower']+b, '--k')
plt.show()
```

<br>Как можно видеть на графике, в целом, с помощью линейной регрессии можно попытаться предсказать, насколько дорогой будет машина, исходя из мощности ее двигателя, но есть очень много отклонений (вбросов) от прямолинейной функции, особенно, по мере увеличения лошадиных сил (правее на графике) - есть машины значительно дороже и дешевле, чем предсказанная стоимость на линейном графике. 
<br>
<br>Теперь для определения коэффициентов линейной функции воспользуемся стандартной функцией из numpy: polyfit

```python
a,b = np.polyfit(data['horsepower'], data['price'], 1)
plt.plot(data['horsepower'], data['price'], 'yo',data['horsepower'], a*data['horsepower']+b, '--k')
plt.show()
```

Теперь воспользуемся средствами оценки точности линейной регрессии (R-квадратом):

```python
r2_score(data['price'], a*data['horsepower']+b)
```

Точность 0.65 хоть и не совсем близка к единице, но тоже говорит о достаточно точно построенной модели
Попробуем добавить параметры для оценки стоимости автомобиля

```python
X = data[['citympg','horsepower','enginesize']]
Y = data['price']
regr = linear_model.LinearRegression()
regr.fit(X, Y)
print(regr.coef_)
predictPrice = regr.predict([[16,182,209]])
print(predictPrice)
```

Мы решили взять расход топлива в городе citympg, мощность horsepower, объем двигателя enginesize
